```python
import pandas as pd
df = pd.read_excel('../test_files/barcode.xlsx', engine = 'openpyxl')
```

```python
df
```

```python
from Bio.Seq import Seq
df['barcode']=df['seq'].str.replace('AGATCGGAAGAGCACACGTC/3AmMO/', '').str.replace('5Phos/NN','').apply(lambda seq: str(Seq(seq).reverse_complement()))
```

```python

df[['RBP', 'barcode']]
```

```python
df['RBP'] = df['RBP'].str.replace(' ', '_')
```

```python
df['barcode_w_name'] = df['barcode']+':'+df['RBP']
```

```python
df['barcode_w_name'].to_csv('../config/barcode_csv/katie_pe.csv',index = False, header = False)
```
