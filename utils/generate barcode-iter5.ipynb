```python
import pandas as pd
from Bio import SeqIO

df = pd.read_csv('/projects/ps-yeolab3/eboyle/encode/pipeline/adapters/katie/barcode_list.tsv', 
                   sep = '\t'
                  ).T.reset_index().T
```

```python
df = df[[0,2]]
```

```python
df.columns = ['name', 'seq']
```

```python
df
```

```python
from Bio.Seq import Seq
df['barcode']=df['seq'].str.replace('AGATCGGAAGAGCACACGTC/3AmMO/', '').str.replace('5Phos/NN','').apply(lambda seq: str(Seq(seq).reverse_complement()))
```

```python
df
```

```python
# MAP RBP TO BARCODE
import pandas as pd
exp_bar = pd.read_csv('../test_files/iter5.csv')[['RBP', 'oligo']].dropna()
```

```python
exp_bar['barcode'] = exp_bar['oligo'].map(df.set_index('name')['barcode'])
```

```python
exp_bar[['RBP', 'barcode']]
```

```python
exp_bar['RBP'] = exp_bar['RBP'].str.replace(' ', '_')
```

```python
exp_bar['barcode_w_name'] = exp_bar['barcode']+':'+exp_bar['RBP']
exp_bar['barcode_w_name'].to_csv('../config/barcode_csv/iter5.csv',index = False, header = False)
```
