```python
import pandas as pd
df = pd.read_excel('../test_files/Barcode_092022_OCLIP.xlsx', engine = 'openpyxl',
                  ).T.reset_index().T
df.columns = ['RBP', 'name', 'seq']
df.reset_index(inplace = True)
```

```python
df
```

```python
from Bio.Seq import Seq
df['barcode']=df['seq'].str.replace('AGATCGGAAGAGCACACGTC/3AmMO/', '').str.replace('5Phos/NN','').apply(lambda seq: str(Seq(seq).reverse_complement()))
```

```python

df[['RBP', 'barcode']]
```

```python
df['RBP'] = df['RBP'].str.replace(' ', '_')
```

```python
df['RBP'] = df['RBP'].str.replace('CHE1/AATF', 'AATF')
```

```python
df['barcode_w_name'] = df['barcode']+':'+df['RBP']
```

```python
df['barcode_w_name'].to_csv('../config/barcode_csv/katie_pe_iter3.csv',index = False, header = False)
```
